```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import pandas as pd
import os

# Define list of your uploaded .sdf filenames
sdf_files = [
    "Cycloheximide_CID_6197.sdf",
    "Dactinomycin_CID_457193.sdf",
    "Lovastatin_CID_53232.sdf",
    "Rifampicin_CID_135398735.sdf",
    "SalinosporamideA_CID_11347535.sdf",
    "Erythromycin_CID_12560.sdf",
    "Fumagillin_CID_6917655.sdf",
    "Staurosporine_CID_44259.sdf",
    "rapamycin_CID_5284616.sdf"
]

# Basic descriptor list
descriptor_funcs = {
    "MW": Descriptors.MolWt,
    "LogP": Descriptors.MolLogP,
    "HBD": Descriptors.NumHDonors,
    "HBA": Descriptors.NumHAcceptors,
    "TPSA": Descriptors.TPSA,
    "Fsp3": Descriptors.FractionCSP3,
    "RotatableBonds": Descriptors.NumRotatableBonds
}

# Store descriptors in a list
data = []

for file in sdf_files:
    suppl = Chem.SDMolSupplier(file)
    mol = suppl[0]  # Get the first conformer from SDF
    if mol is None:
        print(f"❌ Could not read molecule from {file}")
        continue

    # Calculate all descriptors
    descriptors = {desc: func(mol) for desc, func in descriptor_funcs.items()}
    descriptors["Compound"] = file.replace(".sdf", "")
    data.append(descriptors)

# Convert to DataFrame
df = pd.DataFrame(data)
df = df.set_index("Compound")

# Show descriptor table
df
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Crippen, Lipinski
import pandas as pd
import glob

# Path where SDFs are stored (adjust if needed)
sdf_files = glob.glob("*.sdf")  # or "/content/*.sdf" for Colab

results = []

for file in sdf_files:
    suppl = Chem.SDMolSupplier(file)
    mol = suppl[0]
    if mol is None:
        continue

    name = file.replace(".sdf", "")

    mw = Descriptors.MolWt(mol)
    logp = Crippen.MolLogP(mol)
    hbd = Lipinski.NumHDonors(mol)
    hba = Lipinski.NumHAcceptors(mol)
    tpsa = Descriptors.TPSA(mol)
    rot_bonds = Lipinski.NumRotatableBonds(mol)

    lipinski_violations = sum([
        mw > 500,
        logp > 5,
        hbd > 5,
        hba > 10
    ])
    lipinski_pass = lipinski_violations <= 1
    veber_pass = (tpsa <= 140) and (rot_bonds <= 10)

    results.append({
        'Compound': name,
        'MW': round(mw, 2),
        'LogP': round(logp, 2),
        'HBD': hbd,
        'HBA': hba,
        'TPSA': round(tpsa, 2),
        'RotBonds': rot_bonds,
        'Lipinski_Pass': lipinski_pass,
        'Veber_Pass': veber_pass
    })

df = pd.DataFrame(results)
df.sort_values("Compound", inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load molecules
import os
folder = '/content'  # adjust if needed
sdf_files = [
    "Cycloheximide_CID_6197.sdf",
    "Dactinomycin_CID_457193.sdf",
    "Lovastatin_CID_53232.sdf",
    "Rifampicin_CID_135398735.sdf",
    "SalinosporamideA_CID_11347535.sdf",
    "Erythromycin_CID_12560.sdf",
    "Fumagillin_CID_6917655.sdf",
    "Staurosporine_CID_44259.sdf",
    "rapamycin_CID_5284616.sdf"
]

compound_names = [f.split(".")[0] for f in sdf_files]
fps = []

# Generate Morgan fingerprints (radius=2, 2048 bits)
for file in sdf_files:
    suppl = Chem.SDMolSupplier(os.path.join(folder, file))
    mol = suppl[0]
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
    fps.append(fp)

# Compute pairwise Tanimoto similarity
n = len(fps)
sim_matrix = []
for i in range(n):
    row = []
    for j in range(n):
        sim = DataStructs.TanimotoSimilarity(fps[i], fps[j])
        row.append(sim)
    sim_matrix.append(row)

# Convert to DataFrame
df_sim = pd.DataFrame(sim_matrix, index=compound_names, columns=compound_names)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_sim, annot=True, cmap='viridis', fmt=".2f", square=True, cbar_kws={'label': 'Tanimoto Similarity'})
plt.title("Chemical Similarity (Tanimoto) Heatmap")
plt.tight_layout()
plt.show()
```

```python
# ✅ Install RDKit (Colab only; skip if already installed)
!pip install -q rdkit-pypi

# 📦 Imports
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski, AllChem
from rdkit.DataStructs import TanimotoSimilarity
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import dendrogram, linkage

# ✅ List of SDF files (hardcoded)
sdf_files = [
    "Cycloheximide_CID_6197.sdf",
    "Dactinomycin_CID_457193.sdf",
    "Lovastatin_CID_53232.sdf",
    "Rifampicin_CID_135398735.sdf",
    "SalinosporamideA_CID_11347535.sdf",
    "Erythromycin_CID_12560.sdf",
    "Fumagillin_CID_6917655.sdf",
    "Staurosporine_CID_44259.sdf",
    "rapamycin_CID_5284616.sdf"
]

sdf_folder = "/content"  # 👈 Modify if files are elsewhere

compound_names = []
mols = []

# 🔬 Load molecules
for file in sdf_files:
    suppl = Chem.SDMolSupplier(os.path.join(sdf_folder, file))
    mol = suppl[0]
    if mol is not None:
        mols.append(mol)
        name = file.replace(".sdf", "")
        compound_names.append(name)

# 🔗 Generate fingerprints
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048) for mol in mols]

# 🧮 Tanimoto similarity matrix
n = len(fps)
sim_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        sim_matrix[i, j] = TanimotoSimilarity(fps[i], fps[j])

# ✅ Lipinski rule-of-5 check
def passes_lipinski(mol):
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Lipinski.NumHDonors(mol)
    hba = Lipinski.NumHAcceptors(mol)
    return (mw <= 500) and (logp <= 5) and (hbd <= 5) and (hba <= 10)

lipinski_pass = [passes_lipinski(mol) for mol in mols]
labels = [f"{name}\n{'✓' if passed else '✗'}" for name, passed in zip(compound_names, lipinski_pass)]

# 📊 Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sim_matrix, xticklabels=labels, yticklabels=labels, annot=True, fmt=".2f",
            cmap="viridis", square=True, cbar_kws={"label": "Tanimoto Similarity"})
plt.title("Chemical Similarity (Tanimoto) Heatmap with Lipinski Labels")
plt.tight_layout()
plt.savefig("tanimoto_heatmap_lipinski.png", dpi=300)
plt.show()

# 💾 Export similarity matrix
df = pd.DataFrame(sim_matrix, columns=compound_names, index=compound_names)
df.to_csv("tanimoto_similarity_matrix.csv")
print("✅ Saved tanimoto_similarity_matrix.csv")

# 🌿 Optional: Dendrogram
plt.figure(figsize=(8, 6))
dist_matrix = 1 - sim_matrix  # Convert to dissimilarity
linkage_matrix = linkage(squareform(dist_matrix), method="average")
dendrogram(linkage_matrix, labels=compound_names, leaf_rotation=90)
plt.title("Compound Clustering Dendrogram (Tanimoto Distance)")
plt.ylabel("Distance (1 - Tanimoto)")
plt.tight_layout()
plt.savefig("compound_dendrogram.png", dpi=300)
plt.show()
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski, rdMolDescriptors
import pandas as pd
import os

sdf_files = [
    "Cycloheximide_CID_6197.sdf",
    "Dactinomycin_CID_457193.sdf",
    "Lovastatin_CID_53232.sdf",
    "Rifampicin_CID_135398735.sdf",
    "SalinosporamideA_CID_11347535.sdf",
    "Erythromycin_CID_12560.sdf",
    "Fumagillin_CID_6917655.sdf",
    "Staurosporine_CID_44259.sdf",
    "rapamycin_CID_5284616.sdf"
]

descriptor_data = []
for sdf in sdf_files:
    mol = Chem.SDMolSupplier(os.path.join("/content", sdf))[0]
    if mol:
        name = sdf.replace(".sdf", "")
        mw = Descriptors.MolWt(mol)
        logp = Descriptors.MolLogP(mol)
        hbd = Lipinski.NumHDonors(mol)
        hba = Lipinski.NumHAcceptors(mol)
        tpsa = Descriptors.TPSA(mol)
        fsp3 = rdMolDescriptors.CalcFractionCSP3(mol)
        rot_bonds = Descriptors.NumRotatableBonds(mol)
        lipinski = mw <= 500 and logp <= 5 and hbd <= 5 and hba <= 10
        veber = tpsa <= 140 and rot_bonds <= 10

        descriptor_data.append([
            name, mw, logp, hbd, hba, tpsa, fsp3, rot_bonds, lipinski, veber
        ])

# Create dataframe
columns = ["Compound", "MW", "LogP", "HBD", "HBA", "TPSA", "Fsp3", "RotatableBonds", "Lipinski_Pass", "Veber_Pass"]
df = pd.DataFrame(descriptor_data, columns=columns).set_index("Compound")

# Save to CSV
df.to_csv("druglikeness_table.csv")

# Preview
df.head()
```

```python
# Install if needed
!pip install -q pandas matplotlib seaborn scikit-learn scipy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the drug-likeness data table
df = pd.read_csv("druglikeness_table.csv", index_col=0)

# Select descriptors
features = ["MW", "LogP", "HBD", "HBA", "TPSA", "Fsp3", "RotatableBonds"]
X = df[features]
X_scaled = StandardScaler().fit_transform(X)

# Bar plots
for prop in features:
    plt.figure(figsize=(10, 5))
    sns.barplot(x=df.index, y=df[prop])
    plt.xticks(rotation=45, ha='right')
    plt.title(f'{prop} Across Compounds')
    plt.tight_layout()
    plt.savefig(f"barplot_{prop}.png")
    plt.close()

# t-SNE
tsne = TSNE(n_components=2, perplexity=5, random_state=42)
tsne_result = tsne.fit_transform(X_scaled)
tsne_df = pd.DataFrame(tsne_result, columns=["TSNE1", "TSNE2"])
tsne_df["Compound"] = df.index

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tsne_df, x="TSNE1", y="TSNE2", hue="Compound")
plt.title("t-SNE of Compounds")
plt.tight_layout()
plt.savefig("tsne_plot.png")
plt.close()

# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[features].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Descriptors")
plt.tight_layout()
plt.savefig("correlation_matrix.png")
plt.close()

# Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df["KMeans_Cluster"] = kmeans.fit_predict(X_scaled)

agg = AgglomerativeClustering(n_clusters=3)
df["Agglomerative_Cluster"] = agg.fit_predict(X_scaled)

# Dendrogram
linkage_matrix = linkage(X_scaled, method="ward")
plt.figure(figsize=(8, 6))
dendrogram(linkage_matrix, labels=df.index, leaf_rotation=90)
plt.title("Hierarchical Clustering Dendrogram")
plt.tight_layout()
plt.savefig("dendrogram.png")
plt.close()

# Summary report
summary = []
for i, row in df.iterrows():
    lipinski = "passed" if row["Lipinski_Pass"] else "failed"
    veber = "passed" if row["Veber_Pass"] else "failed"
    k_cluster = row["KMeans_Cluster"]
    agg_cluster = row["Agglomerative_Cluster"]
    summary.append(f"{i} {lipinski} Lipinski and {veber} Veber rules; grouped in KMeans cluster {k_cluster} and Agglomerative cluster {agg_cluster}.")

with open("summary_report.txt", "w") as f:
    f.write("\n".join(summary))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def multi_radar_plot(df_norm, features):
    compounds = df_norm.index.tolist()
    angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()
    angles += angles[:1]  # complete the circle

    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    colors = plt.cm.tab10.colors  # Up to 10 colors

    for i, compound in enumerate(compounds):
        values = df_norm.loc[compound, features].astype(float).tolist()
        values += values[:1]  # close loop
        ax.plot(angles, values, label=compound, color=colors[i % 10], linewidth=2)
        ax.fill(angles, values, color=colors[i % 10], alpha=0.1)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(features, fontsize=10)
    ax.set_yticks([0.2, 0.4, 0.6, 0.8])
    ax.set_yticklabels(["0.2", "0.4", "0.6", "0.8"], fontsize=8)
    ax.set_ylim(0, 1)
    plt.title("Radar Chart of All Compounds", y=1.1, fontsize=14)
    plt.legend(loc='upper right', bbox_to_anchor=(1.4, 1.05), fontsize=8)
    plt.tight_layout()
    plt.savefig("multi_compound_radar.png", dpi=300)
    plt.show()

# Call it with:
features = ["MW", "LogP", "HBD", "HBA", "TPSA", "Fsp3", "RotatableBonds"]
multi_radar_plot(df_norm, features)
```
